```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/cancer.csv")
df.head()
```

```python
df.drop(columns=["id", "Unnamed: 32"], inplace = True)
df.head()
```

```python
# Label Endcoding yaitu mengubah data categorical("yang bisa diurutkan") menjadi numerik
# Misalkan jenjang pendidikan :
# SD -> 1
# SMP -> 2
# SMA -> 3
# S1 -> 4

def diagnosis_labeling(diagnosa):
  if diagnosa == "M":
        return 1
  return 0
```

```python
# Apply function ke kolom diagnosis
df["diagnosis"] = df["diagnosis"].apply(diagnosis_labeling)
```

```python
df.head()
```

```python
# Mengubah data menjadi format metrix sekaligus memilih data features dari dataframe
X = df.iloc[:, 1:].values
y = df["diagnosis"].values
```

```python
X # Dalam format Matrix/Array
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.preprocessing import StandardScaler
# Mengubah skla data menjadi lebih kecil
```

```python
# Split data into X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.2,
                                                    random_state = 2023)
```

```python
# Scalling data features --> Mengubah data menjadi skala lebih kecil yang terdistribusi
### Notes --> Hanya dilakukan pada data features

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Design Deep Learning Model Architecture

```python
jumlahKolomData = df.iloc[:, 1:].shape[1]
jumlahKolomData
```

```python
model_1 = tf.keras.models.Sequential([
    #Sequential -> Membuat urutan/rangkaian layer pada model deep learning

    # 3 layer wajib pada model :
    ## Input layer  yaitu Layer dimana data diterima untuk diteruskan ke model
    ## Hidden layer yaitu Layer dimana model mempelajari data yang kita input di input layer
    ## Output Layer yaitu Layer dimana model memberikan output/prediksi dari data yang diterima dan diolah

    tf.keras.layers.Input(shape = (jumlahKolomData)),

    # Untuk Hidden Layer 1
    # Dense Layer (Fully Connected Layer) --> Yaitu Gabungan dari modes/neuron yang berhubungan, dimana mereka menerima input dari lapisan layer sebelumnya untuk diteruskan ke layer selanjutnya

    #Activation Function yaitu rumus matematika yang diaplikasikan pada output dari sebuah layer, ini digunakan agar model bisa mempelajari pola yang lebih kompleks pada data target dengan features
    # Contoh Function :
    # - Sigmoid ==> Logistic Regression rangenya(0,1)
    # - TanH ==> rangenya(-1, 1)
    # - Softmax ==> tidak terbatas
    tf.keras.layers.Dense(15, activation = "relu"),


    # Hidden Layer 2
    tf.keras.layers.Dense(15, activation = "relu"),

    # Output Layer
    tf.keras.layers.Dense(1, activation = "sigmoid")
    # Jumlah nodes/neuron itu menyesuaikan jumlah class pada output


    # Jika kasus classification :
    ### Binary Classification maka activation pada output layer adalah sigmoid
    ### Multi Class Cllasification maka activation pada output layer adalah softmax
])
```

```python
model_1.compile(optimizer = "Adam",
                loss = "binary_crossentropy",
                # Binary Crossentropy  => ketika kasus multi class classification
                # categorical_crossentropy => ketika kasus multi class classification
                metrics = ["accuracy"]) # Bisa juga menggunakan precision, recall

# Optimizer = Parameter yang menyesuaikan paramater" model selama training
# Loss = Menghitung selisih/perbedaan prediksi dengan data asli (test)
# Metrics = Metrik yang digunakan untuk memperbaiki jawaban dari model untuk training selanjutnya
```

```python
# Fase Training

# Goals
# Loss : semakin dilatih maka semakin mengecil
# Accuracy : semakin dilatih maka semakin baik

history_1 = model_1.fit(X_train,
                        y_train,
                        validation_data = (X_test, y_test),
                        epochs = 20) # Epochs yaotu jumlah pengulangan untuk training model
```

```python
# Visualkan dalam diagram model Lossnya

plt.plot(history_1.history["loss"], label = "Loss")
plt.plot(history_1.history["val_loss"], label ="Validation Loss")
plt.legend()
```

```python
plt.plot(history_1.history["accuracy"], label="Accuracy")
plt.plot(history_1.history["val_accuracy"], label = "Validation Accuracy")
plt.legend()
```

# Overfitting

Fenomena yang terjadi pada model jika performa training lebih baik daibandingkan performa testing, Artinya model tidak mempelajari pola hubungan yang ada antar data.

Indikasi Overfitting:
1. Accuracy tinggi, Val accuracy rendah

Sebab terjadi nya Overfitting yaitu:
- Data Training Kurang
- Model terlalu kompleks
-etc

Cara menangani Overfitting ;
- Tambahkan datanya lagi jika kurang
- ubah arsitektur model menjadi lebih sederhana

# Model 2 Overfitting

```python
model_2 = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape = (jumlahKolomData)),
    tf.keras.layers.Dense(15, activation = "relu"),
    # Hidden Layer 2
    tf.keras.layers.Dense(15, activation = "relu"),
    # Output Layer
    tf.keras.layers.Dense(1, activation = "sigmoid")
])
```

```python
model_2.compile(optimizer = "Adam",
                loss = "binary_crossentropy",
                metrics = ["accuracy"]) # Bisa juga menggunakan precision, recall
```

```python
history_2 = model_2.fit(X_train,
                        y_train,
                        validation_data = (X_test, y_test),
                        epochs = 10)
```

```python
plt.plot(history_2.history["loss"], label = "Loss")
plt.plot(history_2.history["val_loss"], label ="Validation Loss")
plt.legend()
```

```python
plt.plot(history_2.history["accuracy"], label="Accuracy")
plt.plot(history_2.history["val_accuracy"], label = "Validation Accuracy")
plt.legend()
```

```python
pred_model2 = model_2.predict(X_test) # Masih dalam format Tensor
pred_model2[0:5]
```

```python
prediction_model2 = np.round(pred_model2).flatten()
prediction_model2
```

```python
print(classification_report(prediction_model2, y_test))
```
